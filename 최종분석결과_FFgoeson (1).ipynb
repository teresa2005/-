### - 2020, 2021 월별 검색량 (부산, 전주, 부천, 제천)

```python
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore',category=UserWarning)
import pandas as pd
import numpy as np
import re  
from datetime import datetime, timedelta,date
# 정규화하기 위해서 =re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

## matplotlib의 한글 문제
import matplotlib.pyplot as plt
%matplotlib inline

# 한글깨짐 해결 
from matplotlib import font_manager, rc 
rc('font', family='HCR Dotum') 
```

```python
# 우선 'utf-8(BOM)'로 저장되어 있는 데이터의 인코딩 형식을 'utf-8로 바꿔준다.'
# 2020-01~2020-12 데이터 불러들이고 병합 
df_2020=pd.DataFrame()
for i in range(1,13):
    if i < 10 :
        text='DM_CLTUR_FSTVL_SCCNT_RESULT_20200'+str(i)+'.csv'
        d=pd.read_csv(text,encoding='utf-8') # 데이터 읽기 
        d['SCCNT_YM']='2020-0'+str(i) # 형태변환 틀에 맞추기 (예:2020-02)
        d['RANK'] = d["SCCNT"].rank(method='dense',ascending=False).astype('int64') # 매달 검색량 순위 변수 생성 
    if i >= 10 : 
        text='DM_CLTUR_FSTVL_SCCNT_RESULT_2020'+str(i)+'.csv'
        d=pd.read_csv(text,encoding='utf-8');d["SCCNT_YM"]='2020-'+str(i)
        d['RANK'] = d["SCCNT"].rank(method='dense',ascending=False).astype('int64')
    df_2020=pd.concat([df_2020,d], axis=0 , ignore_index=False)

# 2021-01~2021-06 데이터 불러들이고 병합 
df_2021=pd.DataFrame()
for i in range(1,7):
    if i < 10 :
        text='DM_CLTUR_FSTVL_SCCNT_RESULT_20210'+str(i)+'.csv'
        d=pd.read_csv(text,encoding='utf-8');d['SCCNT_YM']='2021-0'+str(i) ;d['SCCNT_YM']=pd.to_datetime(d['SCCNT_YM'])
        d['RANK'] = d["SCCNT"].rank(method='dense',ascending=False).astype('int64')
    df_2021=pd.concat([df_2021,d], axis=0 , ignore_index=False)
# df_2020
```

```python
# 2020년도에서 국제영화제에 관련한 검색 데이터만 추출 
mov_2020=df_2020[['SCCNT_YM','SRCHWRD_NM','SCCNT','RANK']]
mov_2020['SCCNT_YM']=pd.to_datetime(mov_2020['SCCNT_YM'])# 날짜 형태로 변환 
mov_2020=mov_2020[mov_2020['SRCHWRD_NM'].str.contains('영화제')]# 영화제 포함한 데이터 추출 
mov_2020=mov_2020[mov_2020['SRCHWRD_NM'].str.contains('국제')]  # 영화제 중 국제 관련 축제만 데이터 추출 
mov_2020=mov_2020[-mov_2020['SRCHWRD_NM'].str.contains('어린이')]  # 관련없는 데이터 삭제
mov_2020=mov_2020[-mov_2020['SRCHWRD_NM'].str.contains('전야제')] 
# mov_2020
```

```python
# 2021년도에서 국제영화제에 관련한 검색 데이터만 추출 
mov_2021=df_2021[['SCCNT_YM','SRCHWRD_NM','SCCNT','RANK']]
mov_2021['SCCNT_YM']=pd.to_datetime(mov_2021['SCCNT_YM'])
mov_2021=mov_2021[mov_2021['SRCHWRD_NM'].str.contains('영화제')];mov_2021=mov_2021[mov_2021['SRCHWRD_NM'].str.contains('국제')]   
mov_2021=mov_2021[-mov_2021['SRCHWRD_NM'].str.contains('어린이')];mov_2021=mov_2021[-mov_2021['SRCHWRD_NM'].str.contains('전야제')] 
#mov_2021
```

```python
# 그래프 그리기 
plt.figure(figsize=(20,10)) ; plt.subplots_adjust(wspace=0.5,hspace=0.3)   
mov_list=mov_2020.SRCHWRD_NM.unique()

for mov in mov_list:
    d_ = mov_2020[(mov_2020['SRCHWRD_NM']==mov)]
    plt.plot(d_.SCCNT_YM, d_.SCCNT, "--o", label=str(mov), markersize=20,alpha=.6) # 그래프 추가
    for i, v in enumerate(d_.SCCNT_YM):
        plt.text(v, list(d_['SCCNT'])[i], list(d_['RANK'])[i],fontsize = 15, color='blue',
                 horizontalalignment='center',verticalalignment='baseline')   
    plt.title('2020 주요 국제 영화제 검색량 변화',size=30); plt.legend(fontsize=20)
    plt.xticks(ticks=d_.SCCNT_YM.to_list(),rotation=45)
    plt.xlabel('○ 안은 검색량 순위''\n'' ※ 순위가 비슷한 검색량은 숫자가 겹쳐짐', size=15)    
    
plt.figure(figsize=(20,10))
for mov in mov_list:
    d_ = mov_2021[(mov_2021['SRCHWRD_NM']==mov)]
    plt.plot(d_.SCCNT_YM, d_.SCCNT, "--o", label=str(mov), markersize=20,alpha=.6) # 그래프 추가
    for i, v in enumerate(d_.SCCNT_YM):
        plt.text(v, list(d_['SCCNT'])[i], list(d_['RANK'])[i],fontsize = 15, color='blue',
                 horizontalalignment='center',verticalalignment='baseline')   
    plt.title('2021 주요 국제 영화제 검색량 변화',size=30);plt.legend(loc=2,fontsize=20)
    plt.xticks(ticks=d_.SCCNT_YM.to_list(),rotation=45)
    plt.xlabel('○ 안은 검색량 순위''\n'' ※ 순위가 비슷한 검색량은 숫자가 겹쳐짐', size=15)    
```

- 매달 개최되는 전국의 모든 문화축제 중, 연중 내내 검색량 상위권을 차지한 주요 국제영화제의 검색량 변화추세를 추출하였습니다.

- 각 영화제는 개최 시기에 월등히 높은 검색량 순위를 차지합니다.


- 특히, '부산 국제영화제'의 경우, 개최 4달 전인 6월부터 개최 2달 후인 12월까지 꾸준히 전체 검색량 순위 5위 내를 지키며 상당히 높은 관심도를 유지하고 있습니다.


- '전주 국제영화제' 또한, 연초부터 개최시기까지 매우 높은 검색량 순위를 유지합니다.

```python
# (kt) 데이터 : 'ANSI'로 인코딩 되어있으므로, 'encoding=cp949' 실시 
j_day_df=pd.read_csv('(KT)플랫폼통합경진대회_전주시_행정동_1.일별관광객수.csv',encoding='cp949')

# I_CNT=현지인 관광색 수,O_CNT=외지인 관광색 수, F_CNT=외국인 관광객 수
# 변수 정보를 보니, 위 3개 변수가 object 문자형으로 지정되어 있음
# 관광객 수가 천단위를 넘어갈때 ',' 쉼표가 찍히고, 0은 '-'으로 되어 있기 때문 
# print(j_day_df.info()) 

# I_CNT,O_CNT, F_CNT 변수에 천단위 ','를 제거하고 숫자형으로 변환 
c_list=['I_CNT','O_CNT','F_CNT']
for i in range(0,3):
    c=c_list[i]
    j_day_df[c]=j_day_df[c].str.replace('-', '0')
    j_day_df[c]=j_day_df[c].str.replace(',', '').astype('int64')
    
# ETL_YMD변수를 날짜형식의 datetime으로 변환 
for i in range(0,len(j_day_df)):
    j_day_df.ETL_YMD[i]=str(j_day_df.ETL_YMD[i])[:4]+'-'+str(j_day_df.ETL_YMD[i])[4:6]+'-'+str(j_day_df.ETL_YMD[i])[6:]
j_day_df.ETL_YMD=pd.to_datetime(j_day_df.ETL_YMD)
# ETL_YMD변수를 인덱스로 지정 #k_day_df=k_day_df.set_index('ETL_YMD')
# print('\n',j_day_df.info()) 
```

### - (코레일) <전주 국제영화제> 기간 관광객 유입 경로 및 특성 분석

```python
# 데이터 불러오기 -  역정보 
d2019=pd.read_csv('고객 특성별 철도서비스 이용정보(강릉,목포,전주)_2019.csv',encoding='utf-8')
d2020=pd.read_csv('고객 특성별 철도서비스 이용정보(강릉,목포,전주)_2020.csv',encoding='utf-8')
d2021=pd.read_csv('고객 특성별 철도서비스 이용정보(강릉,목포,전주)_2021.csv',encoding='utf-8')
#d2019.info() # 운행일자가 int64로 정수 타입 => datetime 형식으로 변환 필요 
```

```python
# 2019-2021 데이터 병합 
data=pd.concat([d2019,d2020],axis=0,ignore_index=False)
data=pd.concat([data,d2021],axis=0,ignore_index=False)
# data
```

```python
# 도착역명이 어떠한 방식으로 할당되어 있는지 파악 
# data. 도착역명.unique()
```

```python
# 도착역이 전주인 데이터 추출 
j_df=data[data['도착역명']=='전주'].reset_index(drop=True) # 도착역명 변수에서 '전주'를 포함한 데이터만 추출
#print(j_df.도착역명.unique()) # 앞선 코드가 잘 실행되었는지 파악 
#print(j_df.도착역명.shape)
#j_df.head()
```

```python
# 전주데이터의 운행일자를 datetime형식으로 변환 
for i in range(0,len(j_df)):
    j_df.운행일자[i]=str(j_df.운행일자[i])[:4]+'-'+str(j_df.운행일자[i])[4:6]+'-'+str(j_df.운행일자[i])[6:]
j_df.운행일자=pd.to_datetime(j_df.운행일자)
# j_df.info()
```

```python
# 전주 데이터를 회원 vs 비회원 데이터로 분리 
j_df_M=j_df[j_df['구분']=='회원'].reset_index(drop=True)
j_df_G=j_df[j_df['구분']=='비회원'].reset_index(drop=True)
#print("비회원 수 => ",j_df_M.shape)
#print("회원 수 => ",j_df_G.shape)
```

```python
# print('성별변수의 고유값 before => ',j_df_M.성별.unique())
# 성별 변수에서 *가진 행 삭제
ss = j_df_M.성별.isin(['*'])
j_df_M=j_df_M[~ss ]
# print('성별변수의 고유값 after => ',j_df_M.성별.unique())
```

```python
# 영화제기간 도착역이 전주일때 남여의 비율 
# 남자와 여자인 경우를 카운트한 데이터프레임 형성 
s_df= j_df_M.groupby(['운행일자','성별'])['도착역명'].count().reset_index()
s_df19=s_df.query("'2019-05-02'<=운행일자 <= '2019-05-11'")
s_df20=s_df.query("'2020-05-28'<=운행일자 <= '2020-05-31'")
s_df21=s_df.query("'2021-04-29'<=운행일자 <= '2021-05-08'")
s_df=pd.concat([s_df19,s_df20,s_df21],axis=0); 
# s_df.head()
```

### 외지인 관광객의 성별 분석

<center> ◇ LSTM(Long Short Term Memory) 알고리즘은 RNN의 기울기 소실 문제를 해결하기 위해서 나온 RNN 알고리즘 종류의 하나로서,<br>
    대표적으로 IoT데이터와 같은 시계열 데이터를 분석할 때 주로 사용됩니다.

### LSTM 모형 적합

- 2019년 1월부터 2021년 5월까지 코레일을 통해 강릉역에서 하차한 인원 수를 예측하고자 합니다. 


- 모형 구축 및 학습 프로그램은 파이썬(Python)의 텐서플로우(Tensorflow)기반의 라이브러리인 케라스(Keras)를 이용하여 구축하였습니다.


- 모형 학습시 mse를 학습의 기준치로 설정하고, optimizer는 adam(Kinfma and Ba, 2014)을 이용하여 모형의 오차값을 줄이고자 합니다. 


- epoch값은 100, batch size는 150으로 고정하였습니다.

```python
# 승하차 데이터 불러오기 
df_2019=pd.read_csv('역별 승하차실적_2019.csv',encoding='utf8')
df_2020=pd.read_csv('역별 승하차실적_2020.csv',encoding='utf8')
df_2021=pd.read_csv('역별 승하차실적_2021.csv',encoding='utf8')
# 병합 
df=pd.concat([df_2019,df_2020,df_2021],axis=0,ignore_index=True)
# 필요한 변수 추출 
df=df[df.정차역=='강릉'];df=df[['운행일자','하차인원수']].reset_index(drop=True)
# 운행일자 변수를 날짜형식으로 변환 
for i in range(0,len(df)):
    df.운행일자[i]=str(df.운행일자[i])[:4]+'-'+str(df.운행일자[i])[6:8]+'-'+str(df.운행일자[i])[10:12]
df.운행일자=pd.to_datetime(df.운행일자)
#하차인원수 정수형으로 변환 
df.하차인원수=df.하차인원수.str.replace(',', '').astype('int64')
# 날짜별데이터로  변환 
df=df.groupby('운행일자').sum().reset_index()
# 하차인원수에 +1한후(log(0)=무한대이므로) log변환 
df.하차인원수=np.log(df.하차인원수+1)
```

```python
# 주기를 사용하는 함수를 이용해 데이터 준비
def split_xy(dataset, time_steps, y_column):
    x,y = list(),list()
    for i in range(len(dataset)):
        x_end_number = i + time_steps
        y_end_number = x_end_number + y_column        
        if y_end_number > len(dataset):
            break            
        tmp_x = dataset[i:x_end_number , :]
        tmp_y = dataset[x_end_number: y_end_number,0]# y값은 0열에 위치한 amount
        x.append(tmp_x)
        y.append(tmp_y)
    return np.array(x), np.array(y)
# 다시 행렬을 구조를 잡아줘야 한다. 
ddd=np.array(df.하차인원수).reshape(df.shape[0],1).astype(np.float)    
x,y = split_xy(ddd,150,1) #  150일 주기로 다음날을 예측하게함
```

```python
# 데이터 나누기 
from sklearn.model_selection import train_test_split
# from sklearn.model_selection import cross_val_score
x_train, x_test, y_train, y_test =train_test_split(
                                 x,y,random_state=1, test_size=0.072, shuffle=False) # 마지막 3달 예측 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras import optimizers

model_lstm=Sequential()
model_lstm.add(LSTM(64, input_shape=(150,1)))
model_lstm.add(Dense(32, activation='relu'))
model_lstm.add(Dense(32, activation='relu'))
model_lstm.add(Dense(32, activation='relu'))
model_lstm.add(Dense(32, activation='relu'))
model_lstm.add(Dense(1)) 
model_lstm.compile(optimizer='Adam', loss='mse', metrics=['mape']) 
```

```python
# shuffle ; False (날짜순서대로) !!! 
np.random.seed(5) 
history_LSTM=model_lstm.fit(x_train,y_train, verbose=1,
                                                 batch_size=150, epochs=100,shuffle=False)
mse,mape=model_lstm.evaluate(x_test,y_test,batch_size=30)
print('mse:',mse)
print('mape:',mape)
```

```python
plt.plot(history_LSTM.history["loss"])
plt.title("Loss")
plt.show() # 손실함수가 감소함 good
```

```python
#train결과
print('-----train-----')
y_pred = model_lstm.predict(x_train)
for i in range(5):
    print('실제:',y_train[i],'/예측:',y_pred[i])
print('\n','-----test-----')
# test결과 
y_test_pred = model_lstm.predict(x_test)
for i in range(5):
    print('실제:',y_train[i],'/예측:',y_test_pred[i])
```

```python
plt.figure(figsize=(20, 10));plt.subplots_adjust(hspace=0.5) 
plt.subplot(2,1,1)
plt.plot(y_train, "b-", label="actual");plt.plot(y_pred, "r-", label="predict")
plt.title('2019~2020년도까지의 강릉에서 하차인원수 : LSTM 학습 결과',size=30); plt.legend(fontsize=20)
plt.subplot(2,1,2)
plt.plot(y_test, "b-", label="actual")
plt.plot(y_test_pred, "r-", label="predict")
plt.title('2021년도 1월부터 5월까지의 강릉에서 하차인원수 : LSTM 테스트 결과',size=30); plt.legend(fontsize=20)
```

```python
a_axis = np.arange(0, len(y_train))
b_axis = np.arange(len(y_train), len(y_train) + len(y_test_pred))
c_axis = np.arange(0, len(y_train) + len(y_test_pred))
y_test_pred=np.array(y_test_pred);y_test=np.array(y_test)
plt.figure(figsize=(20,6))
plt.plot(c_axis, y.reshape(732,),"b-",alpha=0.2, label='Actual')
plt.plot(b_axis, y_test_pred.reshape(53,), '-', color='red', label='Predicted')
plt.plot(a_axis, y_pred.reshape(679,), '-', color='red')
plt.title('2019~2021년도까지의 강릉에서 하차인원수 : LSTM 테스트 결과',size=30)
plt.legend(fontsize=20);plt.show()
```

```python
# LSTM 성능평가 => RMSE
from sklearn.metrics import mean_squared_error
train_RMSE = mean_squared_error(y_train, y_pred) **0.5
test_RMSE = mean_squared_error(y_test, y_test_pred) **0.5
print('train data => RMSE={}'.format(train_RMSE))
print('testdata => RMSE={}'.format(test_RMSE))
```

### LSTM 모형 적합 - 결과 해석

- LSTM모형을 적합한 결과, 위의 그래프에 보이는 바와 같이 시간의 흐름에 따른 하차 인원수가 잘 예측되였습니다. 


- 특히, 예측값들이 저점을 잘 예측하는 특성이 눈에 띕니다. 최소인원을 더욱 잘 예측한다면 추후 활용 가능성이 높아지므로, 이러한 결과는 긍정적으로 해석할 수 있습니다.


- 손실함수 또한, 점차 감소하는 방향으로 0에 수렴하고 있어, 모형이 잘 적합되었음을 알 수 있습니다. 


- 모형 비교를 위한 통계량으로는 RMSE(Root MSE : 평균제곱근 오차)를 하였고, RMSE는 0.4정도로 산출되었습니다. 


- 모형 적합에 사용한 데이터가 적음에도 불구하고, 예상보다 좋은 결과가 나타났습니다. 관련 데이터가 더욱 축적된 상황에서 모델링을 진행한다면, 더욱 발전된 결과를 이끌어낼 수 있을 것입니다.
